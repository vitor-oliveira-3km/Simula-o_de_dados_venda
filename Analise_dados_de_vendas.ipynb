```python
#=> A idea aqui é gerar dados e relatórios referentes a vendas de produtos eletronicos, levando em consideração as cidades brasileiras, quantidade de vendas, Qual produto foi o mais vendido. A intenção também é trabalhar
#=> as habilidades prática da linguagem python e utilizar outros editores de código. ENTÂO VAMOS LÁ .......

#=> ######### Lógica de programação ##########

#=> É nescessario entender as condições para que possa resolver o problema e criar os relatórios

#=> Primeiro ponto para iniciar é percorrer todos os arquivos da pasta de base de dados. Para isso precisamos acessar a lista de dados, nesse caso utilizei uma lista de informações de vendas criadas com o intuito educacional.

#=> Abaixo foi acrescentado as bibliotecas utilizadas na linguagem python. E o endereço do arquivo
import os
import pandas as pd
import plotly.express as px

lista_arquivos = os.listdir("/content/drive/MyDrive/Curso de python/Vendas") #Se fosse local iria passar o endereço que está a pasta (EX: "E://usuario/vitor/downloads/vendas")
display(lista_arquivos)



```

```python
#=> Aqui estabelecimos os primeiros tratamentos de dados
tabela_total = pd.DataFrame()

# Defindo a lista_arquivos
lista_arquivos = os.listdir("/content/drive/MyDrive/Curso de python/Vendas")

#=> importando as bases de dados de vendas
for arquivo in lista_arquivos:
  # se tem vendas no nome do arquivo, eu vou importar o arquivo
  if "Vendas" in arquivo:

    #importa arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Curso de python/Vendas/{arquivo}")
    tabela_total = pd.concat([tabela_total, tabela])
#=> tratar/compilar as bases de dados
display(tabela_total)

#=> Dessa forma é possivel ver a tabela de referência
```

```python
#=> Aqui fazemos os calculos com a finalidade de gerar alguns parametros de comparação, assim como tabelas e gráficos

#=> calcular o produto mais vendido (quantidade)
tabela_produtos = tabela.groupby("Produto").sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending= False)
display(tabela_produtos)



#=> calcular a loja/cidade que mais vendeu (faturamento)- e criar uma dashboard/gráfico
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']


tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending= False)
display(tabela_faturamento)




# por EX: if condição > 1000
```

```python

#=> calcular a loja/cidade que mais vendeu (faturamento)-criar uma dashboard/gráfico

tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']]
display(tabela_lojas)
```

```python
#=> Aqui geramos um grafico em barra um pouco mais detalhado referente a cidade que mais faturou, Sendo facil extender para qualquer parâmetro nescessário
grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y= 'Faturamento')
grafico.show()
```
